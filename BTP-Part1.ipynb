## **Libraries Used**

```
import numpy as np
import pandas as pd
import nltk
from nltk import *
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
from wordcloud import WordCloud
from tabulate import tabulate
import string
import matplotlib.pyplot as plt 
import re
lemmatizer = WordNetLemmatizer()
stopwords = stopwords.words("english")
from nltk import FreqDist
```

## **Pre-processing**

```
#data preprocessing
def to_lowercase(text):
    return  text.lower()
def remove_numbers(text):
    result = re.sub(r'\d+', '', text)
    return result
def remove_punctuation(text):
    text = re.sub(r'[“|”|‘|’]',' ',text)
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)
def remove_whitespace(text):
    return " ".join(text.split())
def lemmatize_word(text):
    word_tokens = word_tokenize(text)
    lemmas = [lemmatizer.lemmatize(word, pos='v') for word in word_tokens]
    return lemmas

file1 = open('T11.txt','r')
file2 = open('T22.txt','r')
text1 = file1.read()
text2= file2.read()


text1 = to_lowercase(text1)
text1 = remove_numbers(text1)
text1 = remove_punctuation(text1)
text1 = remove_whitespace(text1)
text1_tokenized = lemmatize_word(text1)

print("\n\nNumber Of Tokens in T1->",len(text1_tokenized))
print("Number of Vocab in T1->",len(set(text1_tokenized)))

text2 = to_lowercase(text2)
text2 = remove_numbers(text2)
text2 = remove_punctuation(text2)
text2 = remove_whitespace(text2)
text2_tokenized = lemmatize_word(text2)

print("\n\nNumber Of Tokens in T2->",len(text2_tokenized))
print("Number of Vocab in T2->",len(set(text2_tokenized)),"\n\n")
```
